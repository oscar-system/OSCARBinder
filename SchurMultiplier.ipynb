# 1 Computing the Schur Multiplier

Issai Schur showed that every finite group G has associated to it at least one finite group $C$ (called a Schur cover) with the property that every projective representation of $G$ can be lifted to an ordinary representation of $C$. Schur covers are special instances of central extensions, i.e. they satisfy a short exact sequence

$$
1 \to M \to C \xrightarrow{f} G \to 1
$$

where $M \cong H_2(G, \mathbb{Z})$. Consequently, we can compute the Schur multiplier $H_2(G, \mathbb{Z})$ as the kernel of the map from a Schur cover $C$ to $G$.

### Examples

```julia
G = cyclic_group(5)
schur_multiplier(G)
```

The Schur multiplier is trivial for cyclic groups.

```julia
is_trivial(schur_multiplier(cyclic_group(5)))
```

Of course, in general the Schur muliplier is nontrivial:

```julia
G = symmetric_group(3)
schur_multiplier(G)
```

```julia
G = symmetric_group(4)
schur_multiplier(G)
```

```julia
is_trivial(schur_multiplier(G))
```

# 2 Restricting to Subgroups

Given a subgroup $U \subseteq G$, an element in $H_2(G, \mathbb{Z})$ can be restricted to $H_2(U, \mathbb{Z})$. Some elements may:
- be nontrivial globally,
- but become trivial when restricted to subgroups.

We now compute the subgroup of $H_2(G, \mathbb{Z})$ consisting of elements that become trivial when restricted to a family of subgroups.

```julia
function restriction_kernel(G, U)
    _, f = schur_cover(G)
    k = kernel(f)[1]
    im_gen = elem_type(k)[]
    for D in U
        gp = intersect(k, derived_subgroup(preimage(f, D[1])[1])[1])[1]
        for g in gens(gp)
            push!(im_gen, g)
        end
    end
    n = sub(k, im_gen)[1]
    n = normal_closure(k, n)[1]
    return quo(k, n)
end
```

### Example: symmetric group $S_4$

```julia
G = symmetric_group(4)
schur_multiplier(G)
```

```julia
U = [sylow_subgroup(G, 2), sylow_subgroup(G, 3)]
```

```julia
restriction_kernel(G, U)
```
